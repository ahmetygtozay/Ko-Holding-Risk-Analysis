# Veri Analizi

```python
import pandas as pd
import matplotlib.pyplot as plt

# Verilerinizi içeren bir DataFrame oluşturun
# 'Date' sütunu tarihleri, 'Close' sütunu kapanış fiyatlarını içermelidir
# Verileriniz bu örnekle aynı şekilde düzenlenmelidir
data = {
    'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05'],
    'Close': [13289345, 13955194, 14083888, 13459734, 14310459]
}

df = pd.DataFrame(data)

# Günlük getiri oranlarını hesaplayın
df['Daily_Return'] = df['Close'].pct_change() * 100

# Volatilite grafiğini oluşturun
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Daily_Return'], marker='o', linestyle='-')
plt.title('Günlük Volatilite Grafiği')
plt.xlabel('Tarih')
plt.ylabel('Günlük Değişim (%)')
plt.grid(True)
plt.xticks(rotation=45)  # Tarih etiketlerini 45 derece döndürün

# Grafiği gösterin
plt.show()
```

# Risk Analizi Tablosu

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Verileri bir DataFrame'e dönüştürün
data = {
    'Finansal Varlık': ['Koç Holding'],
    'Getiri': [0.03],
    'Ortalama Getiri': [0.02],
    'Standart Sapma': [0.05],
    'Varyans': [0.0025],
    'Kovaryans': [0.001],
    'Beta Katsayısı': [1.2],
    'Minimum Getiri': [0.01],
    'Maksimum Getiri': [0.05],
    'Risk/Gelir Oranı': [2.5]
}

df = pd.DataFrame(data)

# Veri çerçevesini dönüştürün
df = df.set_index('Finansal Varlık').T

# Çubuk grafik oluşturun
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")  # Arkaplanı beyaz yapın
sns.barplot(x=df.index, y=df['Koç Holding'], hue=df.index, palette="viridis", legend=False)
plt.xticks(rotation=45)  # Metrik adlarını 45 derece döndürün
plt.title('Risk Analizi - Koç Holding')
plt.xlabel('Metrik')
plt.ylabel('Değer')

# Grafiği gösterin
plt.show()
```
